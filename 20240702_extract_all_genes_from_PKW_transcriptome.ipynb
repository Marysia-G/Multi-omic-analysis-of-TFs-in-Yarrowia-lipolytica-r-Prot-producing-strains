```python
import pandas as pd

pd.options.display.max_rows = 5000

# Load the Excel file
xls = pd.ExcelFile(r'C:\Users\User\Desktop\PROJEKTY_VS\ICY24-CapeTown\transcriptomics20240424\PKW-Supplement-Global_transcriptome_profiling.xlsx')

# Dictionary to store each sheet's DataFrame
dfs = {}

# Loop through each sheet and collect the specified columns
for sheet_name in xls.sheet_names:
    df = pd.read_excel(xls, sheet_name=sheet_name)
    if 'Gene' in df.columns and 'log2_Estimated_FoldChange' in df.columns:
        dfs[sheet_name] = df[['Gene', 'log2_Estimated_FoldChange']]
'''
# Display the extracted data
for sheet_name, data in dfs.items():
    print(f"Sheet: {sheet_name}")
    print(data.head())
    print()'''

# Now, merge all the DataFrames on the 'Gene' column
merged_df = pd.DataFrame()
for sheet_name, df in dfs.items():
    df = df.rename(columns={'log2_Estimated_FoldChange': f'log2_estimated_fold_change_{sheet_name}'})
    if merged_df.empty:
        merged_df = df
    else:
        merged_df = pd.merge(merged_df, df, on='Gene', how='outer')

# Display the merged DataFrame
print(merged_df)
```

```python
xls2 = pd.ExcelFile(r'C:\Users\User\Desktop\PROJEKTY_VS\ICY24-CapeTown\transcriptomics20240424\PKW-Supplement-Molecular_background_HAC1.xlsx')

# Dictionary to store the extracted data
extracted_data2 = {}

# Loop through each sheet and collect the specified columns
for sheet_name in xls2.sheet_names:
    df2 = pd.read_excel(xls2, sheet_name=sheet_name)
    if 'MAPPED IDs' in df2.columns and 'log2_Estimated_FoldChange' in df2.columns:
        extracted_data2[sheet_name] = df2[['Gene', 'log2_Estimated_FoldChange']]

# Display the extracted data
for sheet_name, data2 in extracted_data2.items():
    print(f"Sheet: {sheet_name}")
    print(data2.head())
    print()
```

```python
merged_df2 = pd.concat(extracted_data2, ignore_index=True)
print(merged_df2)
```

```python
#All genes in transcriptomics data

all_genes = pd.merge(merged_df, merged_df2, on='Gene', how='outer')
all_genes.rename({'log2_estimated_fold_change_SoA_all_genes': 'scSoA',
                  'log2_estimated_fold_change_scYFP_all_genes': 'scYFP',
                  'log2_estimated_fold_change_inYFP_all_genes': 'inYFP',
                  'log2_estimated_fold_change_TlG_all_genes': 'scTlG',
                  'log2_Estimated_FoldChange': 'HAC_scYFP'}, axis=1, inplace=True)
all_genes.fillna(0, inplace=True)
print(all_genes)
```

```python
all_genes.to_excel('all_transcriptomics_genes_PKW.xlsx')
```
